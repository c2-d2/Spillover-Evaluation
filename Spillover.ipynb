```python
from google.colab import files
import pandas as pd
import re

# Step 1: Upload the file
uploaded = files.upload()

# Load the CSV file
df = pd.read_csv('SpilloverRankings.csv')

# Check the existing columns
print("Columns in the dataframe:", df.columns)
```

```python
# Column containing the complex data
data_column = 'Risk Levels'

# Function to extract information from each cell
def extract_info(cell):
    pattern = r'Risk Name: (.+?)\nRisk Score: ([0-9.]+)\nImpact Score: ([0-9.]+)\nWeighted Score: ([0-9.]+)\nCorresponding level/levels: ([^\n]*)(?:\n|$)'
    matches = re.findall(pattern, cell, re.MULTILINE)
    return matches

# Apply the extraction function to each row in the "Risk Levels" column
extracted_data = df[data_column].apply(extract_info)

# Debug: print extracted data to verify
print("Extracted data sample:")
print(extracted_data.head())

# Initialize a dictionary to store new columns data
new_columns_data = {}

# Populate the new columns dictionary with extracted data
for idx, entry in extracted_data.items():
    for match in entry:
        risk_name, risk_score, impact_score, weighted_score, _ = match  # Ignore corresponding_levels
        risk_name = risk_name.strip()
        # Initialize columns if they don't exist
        if f'{risk_name} - Risk Score' not in new_columns_data:
            new_columns_data[f'{risk_name} - Risk Score'] = [None] * len(df)
            new_columns_data[f'{risk_name} - Impact Score'] = [None] * len(df)
            new_columns_data[f'{risk_name} - Weighted Score'] = [None] * len(df)
        # Populate columns
        new_columns_data[f'{risk_name} - Risk Score'][idx] = risk_score.strip()
        new_columns_data[f'{risk_name} - Impact Score'][idx] = impact_score.strip()
        new_columns_data[f'{risk_name} - Weighted Score'][idx] = weighted_score.strip()

# Debug: print new columns data to verify
print("New columns data sample:")
for column_name, column_data in new_columns_data.items():
    print(f"{column_name}: {column_data[:5]}")  # Print first 5 values for each new column

# Create new columns in the dataframe
for column_name, column_data in new_columns_data.items():
    df[column_name] = column_data

# Save the structured data to a new CSV file
output_file = 'structured_spillover_rankings.csv'
df.to_csv(output_file, index=False)

print(f"Data successfully written to {output_file}")

# Download the structured CSV file
files.download(output_file)
```
